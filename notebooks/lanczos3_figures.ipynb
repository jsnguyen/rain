```python
x0, y0 = 0, 0
amplitude = 1
radius = 1
n_interp = 100
x = np.linspace(-1, 1, n_interp)
y = np.linspace(-1, 1, n_interp)

xs, ys = np.meshgrid(x,y)

vec_lanczos3 = np.vectorize(lanczos3)

res = vec_lanczos3(xs, ys, amplitude, x0, y0, radius)

fig, ax = plt.subplots(figsize=(4,4))

im = ax.imshow(res, vmin=-0.2, vmax=1)
cbar = fig.colorbar(im, ax=ax)

ax.set_title('Lanczos 3 Kernel')

plt.savefig('lanczos3.png', bbox_inches='tight', dpi=300)
plt.show()
```
